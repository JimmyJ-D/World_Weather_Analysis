```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np #remember we don't have to use random because np has it alread
```

```python
#they need to be stored so that we can access them later. 
#Since we are creating arrays of latitudes and longitudes, we'll declare each array as a variable.
```

```python
#In addition, we'll pack the latitudes (lats) and longitudes (lngs) 
#as pairs by zipping them (lat_lngs) with the zip() function.
#NOT SURE ABOUT THE ZIP() FUNCTION it appears to pack them as a matching pair in a tuple index 
#of 0 and 1 respectively

# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs #this the zip of lats & long generated 1500 times randomly

```

```python
#let's unpack our lat_lngs zip object into a list. 
#This way, we only need to create a set of random latitudes and longitudes once.

# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
#You can only unzip a zipped tuple once before it is removed from the computer's memory. 
#Make sure you unzip the latitudes and longitudes into the coordinates list before moving on
```

```python
#6.1.5 Generate Random World Cities 
# installing  pip install citipy
#https://github.com/wingchen/citipy/blob/master/README.md

# Use the citipy module to determine city based on latitude and longitude.
#for me to get this to work I had to import it in (base) not Python like the books said
from citipy import citipy

```

```python
#In a new cell, create a for loop that will do the following:

#Iterate through the coordinates' zipped tuple.
#Use citipy.nearest_city() and inside the parentheses of nearest_city(), 
#add the latitude and longitude in this format: coordinate[0], coordinate[1].
#To print the city name, chain the city_name to the nearest_city() function.
#To print the country name, chain the country_code to the nearest_city() function.


# Use the tuple() function to display the latitude and longitude combinations.
for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
          citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
#THE INSTRUCTION SAID WE SHOULD GET 5 CITIES BUT OUR TUBLE ZIP SIZE IS 1500... ???
```

```python
# Create a list for holding the cities.
cities = [] #CREATE cities list to store city names
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates: #iterate through the coordinates, retrieve the nearest city using lat/long pair
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities: 
        #decision statement with the logical operator "not in" determine wheter the found city is already in the cities list
        #if not we'll use the append() function to add it this make use unique cities only??
        cities.append(city)
# Print the city count to confirm sufficient count.
len(cities)
```
